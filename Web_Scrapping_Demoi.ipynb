```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
for i in soup.find_all('h1'):
  print(i.text)
```

```python
soup.find_all('a')
```

```python
company=soup.find_all('a')
```

```python
len(company)
```

```python
name = []
location = []

companies = soup.find_all('a', class_='eachSimilarBusiness')

for c in companies:
    title = c.find('div', class_='eachSimilarTitle')
    loc = c.find('div', class_='eachSimilarLocation')


    name.append(title.text.strip() if title else '')
    location.append(loc.text.strip() if loc else '')

print(name)
print(location)
```

```python
df=pd.DataFrame({'name':name,'location':location})
```
